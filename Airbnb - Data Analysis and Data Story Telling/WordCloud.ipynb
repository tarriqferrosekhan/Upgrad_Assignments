```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
#reference: https://www.geeksforgeeks.org/generating-word-cloud-python/
def getWordCloud(df,col_name,_title):
    comment_words = ''
    stopwords = set(STOPWORDS)
    _listsource=list(df[col_name].unique())
    for val in _listsource:
        val = str(val)
        tokens = val.split()
        for i in range(len(tokens)):
            tokens[i] = tokens[i].lower()
        comment_words += " ".join(tokens)+" "
    wordcloud = WordCloud(width = 800, height = 800,
                    background_color ='white',
                    stopwords = stopwords,
                    min_font_size = 10).generate(comment_words)
    plt.figure(figsize = (8, 8), facecolor = None)
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.tight_layout(pad = 0)
    _fontdict={
        'color':  'darkred',
        'weight': 'bold',
        'size': 20,
        }
    plt.title(_title,fontdict=_fontdict)
    plt.show()
```

```python
def drawWordCloud(df,room_type,neighbourhood_group):
    _title=room_type+"-"+neighbourhood_group
    _filter_entire_manhattan=((df['room_type']==room_type)& (df['neighbourhood_group']==neighbourhood_group))
    #print(len(df[_filter_entire_manhattan]))
    return getWordCloud(df[_filter_entire_manhattan],'name',_title)
```

```python
df = pd.read_csv(r"AB_NYC_Analysed.csv", encoding ="latin-1")
```

```python
drawWordCloud(df,'Entire home/apt','Queens')
```
